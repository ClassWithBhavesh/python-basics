# ***Vectorization in NumPy***

- Performing operations on entire array at once without using Python loops

- Why use Vectorization in NumPy?
    - Eliminates python loops
    - Uses optimized C-Level inplementation
    - Much Faster as compared to python loops
    - easy to understand code (cleaner code)
    - Increases the Readability of the code

```python
import time as tl

# startTime = tl.time()
# li = [14,15,16,78]
# newli = []
# newLi = li + 5    # [14,15,16,78] + [5, 5 ,5 , 5] --> it wil not gonna apply broadcasting as well as vectorization on python list
# print(newLi)
# through this we can't operate the li with adding 5 to entire python list
# to operate the above operation in python we have to use python loops
# for num in li:
#   newli.append(num + 5)
# endTime = tl.time()

startTime = tl.time()
li = []
newLi = []
for num in range(1, 10000000):
  li.append(num)

for num in li:
  newLi.append(num + 5)

endTime = tl.time()

# print(li)
# print(newLi)
# print(newli)
print(endTime - startTime)
```
