# ***Pre-Built Methods in NumPy for the Numerical Manipulation***

```python
import numpy as np

arr = np.array([[1,2,3], [5,6, 7]])
# arr = np.array([[1,2, 3], [5,6]])   # it will cause an inhomogeneous shape error (ValueError) because the number of columns must be same in every list
print(arr)
print(np.ndim(arr))  # it will return the dimension of the array (1D | 2D | 3D)
print(np.shape(arr)) # it will return the no. of rows and columns (m,n)
print(np.size(arr))  # it will return the count of the elements present in an array
```

# ***Type Conversion in NumPy***

```python
# let us say we have a stringular array -
strArr = np.array(['10', '20', '30', '40'])
print(strArr)
print(strArr.dtype)
# converting stringular array into int array
intArr = np.array(strArr, dtype = np.int64)
print(intArr)
print(intArr.dtype)

# converting stringular array into float array
intArr = np.array(strArr, dtype = np.float64)
print(intArr)
print(intArr.dtype)
```

```python
# np.arange(startRange, endRange, skipValue) - it will make a array of the provided range

rangedArr = np.arange(20, 0, -2)
print(rangedArr)
```

```python
# np.linspace(startPoint, endPoint, divisions)
eqDivArr = np.linspace(0, 20, 10, dtype=np.int64)
print(eqDivArr)
print(np.size(eqDivArr))
```
