# ***Boolean Masking in NumPy***

```python
arr = np.arange(1, 41)
print(arr)

evenArr = arr[arr % 2 == 0]
print(evenArr)

oddArr = arr[arr%2 == 1]
print(oddArr)

# Corrected: Use the bitwise '&' operator for element-wise logical AND
ArrGreaterThan18 = arr[(arr > 18) & (arr < 25)]
print(ArrGreaterThan18)
```

```python
# Demonstrate .any() and .all()

print("\n--- Demonstrating .any() and .all() ---")

# Check if ANY element in arr is greater than 30
is_any_greater_than_30 = (arr > 30).any()
print(f"Is any element in arr greater than 30? {is_any_greater_than_30}")

# Check if ALL elements in arr are greater than 0
is_all_greater_than_0 = (arr > 0).all()
print(f"Are all elements in arr greater than 0? {is_all_greater_than_0}")

# Check if ALL elements in arr are less than 20 (this should be False)
is_all_less_than_20 = (arr < 20).all()
print(f"Are all elements in arr less than 20? {is_all_less_than_20}")
```
