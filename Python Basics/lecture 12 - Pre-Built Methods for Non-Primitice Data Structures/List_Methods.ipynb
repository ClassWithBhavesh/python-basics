# ***Built-In Methods for List in Python***


```python
products = ["Maggi", "Bottle", "Oats", "Paneer", "Soya Chunks", "Coca Cola", "Wild Stone Perfume"]
priceList = [30, 65, 125, 45, 78, 45, 220]
```

```python
# # 1. sort()
print("------------------- Sorting in stringular list ---------------------\n")
print(f"before sorting - {products}\n")
products.sort()
print(f"\nafter sorting - {products}\n")
print("------------------- Sorting in numerical list in ascending order ---------------------\n")
print(f"before sorting - {priceList}\n")
priceList.sort()         # by default it will sort list in ascending order
print(f"\nafter sorting - {priceList}\n")
print("------------------- Sorting in numerical list in descending order ---------------------\n")
print(f"before sorting - {priceList}\n")
priceList.sort(reverse=True)         # Now, it will sort list in descending order
print(f"\nafter sorting - {priceList}\n")
```
