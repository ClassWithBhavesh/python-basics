# ***Memebership Operators in Python***

is used to check whether an element is present in a dataset or not, which means that this operator will return a boolean value by comparing that a value is present in a dataset or not

1.   in
2.   not in

```python
# before practicing memebership and identity operator we must know all Data-Types
# Data-types - type of data which is stored in a variable
# types of data-types -
    #  1. primitive data types - int, float, str, bool, null, complex number
    #  2. non-primitive data types - list, tuple, set, dictionary (dict)
```

```python
tup = (10,40,50,90,30,70,70,85,95,80,45)
# 'in' membership operator in python
# print(70 in tup and 85 in tup and 95 in tup)     # it will compare 75, 85 and 95 with each value of the tup and will return true if 75, 85 and 95 actually exists in tup otherwise it will return false
# print(70 in tup)                                 # it will compare 70 with each value of the tup and will return true if 70 actually exists in tup otherwise it will return false


# 'not in' membership operator in python
# print(30 not in tup)                             # it will return true if the condition is satisfied and will return false if the condition is unsatisfied
# print(37 not in tup)                             # it will return true if the condition is satisfied and will return false if the condition is unsatisfied
```
