# ***List in Python***

*   it is a well defined collection of elements of similar data-types or different data-types
*   it is denoted by '[value1, value2, value3, value4, ..., valueN]'
*   Each and Every Element are separated by commas ','
*   it is mutable data structure (we can change or update the data or values in the list.)
*   it will store data in sequential (Structured manner), that's why it is a well defined (means well structured)
*   Each and Every Elements have their unique identity known as "Index" or "Indices"

# ***Tuple in Python***



*   is a well defined collection of similar data-types elements as well as different data-types elements
*   Now, you'll think that the list and tuples are same. it's not, tuple is slightly different from List
*   List is Mutable (can be change or update anytime), but tuple is immutable (can't be update or changed after declared)
*   it is denoted by '()' parenthetsis
*   each and every element are separated by ',' comma
*   this is a structured (Ordered) non-primitive data-type / structure
*   every element of tuple has a unique identity known as index value or indices
*   index value starts from 0 to (length of tuple - 1)
