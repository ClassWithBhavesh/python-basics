```python
# num1 = 21
# num2 = 86
# result = num1 + num2
# print(result)


# num1 = 21
# num2 = 86
# result = num1 + num2
# print(result)


# num1 = 21
# num2 = 86
# result = num1 + num2
# print(result)


# num1 = 21
# num2 = 86
# result = num1 + num2
# print(result)


# num1 = 21
# num2 = 86
# result = num1 + num2
# print(result)


# num1 = 21
# num2 = 86
# result = num1 + num2
# print(result)


# to declare a function - syntax to use a function
# def funcName():
  # function body (jo bhi operation perform karna hai uska code)

# funcName()

# steps to execute a function - 1. Declaring a function     2. Calling a function

# Types of function -
# 1. Standard function -
# def sum():
#   num1 = 21
#   num2 = 86
#   result = num1 + num2
#   print(result)

# sum();
# sum();
# sum();
# sum();
# sum();
# sum();



# 2. Parametric functions -
# def sum(num1, num2, num3, num4, num5):
#   result = num1 + num2 + num3 + num4 + num5
#   print(result)

# def sum(num1, num2):
#   result = num1 + num2
#   print(result)

# sum(25, 87);
# sum(21, 46);
# sum(-25, 0);
# sum(19, -71);
# sum(-58, 39);
# sum(.0012, 8.96);



# 3. function with return value -
# kisi din kisi situation me aapko function use to karna hai, uska operation to perform karna hai lekin jo resultant value hai usse execute nahi karna hai.
# esi situation me aap usse return kar sakte ho! kya hoga return karne se?
# return karne se jo bhi resultant value hai wo store ho jayegi (return gho jayegi) kiske andar? --> function ke andar jisme wo resultant value operate ki gayi ho

# prepare a full fledged result of 10th standard student whose name is "Kartik"
mathsMarks = int(input("enter Maths marks -"))
scMarks = int(input("enter Science marks -"))
hnMarks = int(input("enter Hindi marks -"))
engMarks = int(input("enter English marks -"))
sstMarks = int(input("enter SST marks -"))
totalMarks = int(input("Enter Total Marks - "))

def obtMarks(mathsMarks, scMarks, hnMarks, engMarks, sstMarks):
  obt_marks = mathsMarks + scMarks + hnMarks + engMarks + sstMarks
  return obt_marks


def percentage(ob_marks, t_marks):
  per = (ob_marks/t_marks) * 100
  return per

def gradingSystem(std_ob_marks, std_per, t_marks):
  if std_per > 90 and std_per <= 100:
    print(f"Kartik scored {std_ob_marks} out of {t_marks} and got {std_per}% and secured 'A' grade")
  elif std_per > 80 and std_per <= 90:
    print(f"Kartik scored {std_ob_marks} out of {t_marks} and got {std_per}% and secured 'B' grade")
  elif std_per > 70 and std_per <= 80:
    print(f"Kartik scored {std_ob_marks} out of {t_marks} and got {std_per}% and secured 'C' grade")
  elif std_per > 60 and std_per <= 70:
    print(f"Kartik scored {std_ob_marks} out of {t_marks} and got {std_per}% and secured 'D' grade")
  elif std_per > 33 and std_per <= 60:
    print(f"Kartik scored {std_ob_marks} out of {t_marks} and got {std_per}% and secured 'E' grade")
  else:
    print("Kartik got failed, better luck for the next time")



ob_marks = obtMarks(mathsMarks, scMarks, hnMarks, engMarks, sstMarks)
std_percentage = percentage(ob_marks, totalMarks)
gradingSystem(ob_marks, std_percentage, totalMarks)
```
