# ***Comparison Operators***


1.   "==" --> equats to
2.   "!=" --> not equals to
3.   ">"  --> greater than
4.   "<"  --> less than
5.   ">=" --> greater than or equals to
6.   "<=" --> less than or equals to



```python
# equals to - '=='
num1 = 154
num2 = 154

print(num1 == num2)  # if both operands are equal than return True otherwise it will return False
```

```python
# "!=" --> not equals to

num1 = 154
num2 = 154.02

print(num1 != num2)  # if both operands are equal than return False otherwise it will return True
```

```python
# ">" --> greater than
# num1 = 14785
# num2 = 154468999999

# print(num1 > num2)
```

```python
# "<" --> less than
num1 = '1525'
num2 = '1544'

print(num1 < num2)
```

```python
# ">=" --> greater than or equals to
num1 = 1544
num2 = 1544

print(num1 >= num2)
```

```python
# "<=" --> less than or equals to
num1 = 13631
num2 = 165

print(num1 <= num2)
```
