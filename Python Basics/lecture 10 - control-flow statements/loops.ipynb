```python
num1 = int(input("Enter first value - "))
num2 = int(input("Enter second value - "))
num3 = int(input("Enter third value - "))

# ascending order - lowest to highest
# num3, num1, num2
# descending order - highest to lowest
# num2, num1, num3

if num3 < num2 and num3 < num1:
  if num2 < num1:
    print("num3, num2, num1")
  elif num1 < num2:
    print("num3, num1, num2")
  else:
    print("num1 and num2 are equal! please provide unique values.")
elif num2 < num1 and num2 < num3:
  if num1 < num3:
    print("num2, num1, num3")
  elif num3 < num1:
    print("num2, num3, num1")
  else:
    print("num1 and num3 are equal! please provide unique values.")
elif num1 < num2 and num1 < num3:
  if num2 < num3:
    print("num1, num2, num3")
  elif num3 < num2:
    print("num1, num3, num2")
  else:
    print("num2 and num3 are equal! please provide unique values.")
else:
  print("All numbers are equal! please provide unique values.")
```
