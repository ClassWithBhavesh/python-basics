```python
import matplotlib.pyplot as plt
```

# **Scatter Plot in Matplotlib**

```python
study_hours = list(range(1, 11))
exam_marks = [35, 48, 52, 58, 65, 68, 75, 81, 88, 92]

plt.plot(study_hours, exam_marks, color='black', zorder=-1, linewidth=1, alpha=1)
# plt.scatter(study_hours, exam_marks, marker='v', color='yellow', edgecolors='darkcyan', s=100)
# plt.scatter(study_hours, exam_marks, marker='v', c=['violet', 'yellow', 'green', 'red', 'purple', 'gray', 'blue', 'orange', 'skyblue', 'brown'], edgecolors='darkcyan', s=100)
plt.scatter(study_hours, exam_marks, marker='v', c=exam_marks, cmap='berlin', edgecolors='darkcyan', s=100)
plt.grid(c='g', alpha=.1)
plt.colorbar(label='marks level indicater')
plt.xlabel('study hours')
plt.ylabel('exam marks')

plt.show()

```

# **Pie Charts in Matplotlib**

```python
prodCat = ['Electronics', 'Accessories', 'Clothing', 'Groceries', 'Crockery', 'Furniture']
sales=[240000, 370000, 690000, 850000, 430000, 960000]

plt.pie(sales, labels=prodCat, labeldistance=1.1, counterclock=False, textprops={'fontsize': '10', 'color': 'yellow', 'backgroundcolor': 'brown'}, colors=['violet', 'blue', 'yellow', 'green', 'purple', 'orange'], autopct='%1.0f%%', pctdistance=0.8, explode=[0,0,0.1,0,0,0.1], startangle=135, shadow=True, wedgeprops={'edgecolor': 'black', 'linewidth': 1, 'width': .6})
plt.title('Sales Distribution / Category')
plt.show()
```
