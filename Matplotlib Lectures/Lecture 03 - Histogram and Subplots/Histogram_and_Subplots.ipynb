# **Subplots In Matplotlib**

```python
match_year = [2014, 2015, 2016, 2017, 2018]

virat_scoring_century = [10, 12, 6, 12, 9]
rohit_scoring_century = [9, 8, 13, 18, 2]

overs = [10,12,8,13,18]

fig, ax = plt.subplots(2, 2, figsize=(15,10))

ax[0,0].plot(match_year, virat_scoring_century, label="Virat's Score", marker='o')
ax[0,0].plot(match_year, rohit_scoring_century, label="Rohit's Score", linestyle='--', marker='v')
ax[0,0].legend()

ax[1, 1].plot(overs, virat_scoring_century, color='violet', marker='d', label="Virat's Score")
ax[1, 1].bar(overs, rohit_scoring_century, color='yellow', edgecolor='black', label="Rohit's Score")
ax[1, 1].legend()

plt.show()
```
