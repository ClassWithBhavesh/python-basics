```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [7700, 2500, 1400, 3400, 2300, 7700]

plt.plot(x, y, color='magenta', linestyle='-', marker='s', label='lap wise distance covered')
plt.legend(loc="lower right", fontsize=7)
plt.title("Lap-Wise Biker Performance")
plt.xlabel("Laps")
plt.ylabel("Distance")

plt.show()
```

# **Line Plotting / Line Graph**

```python
time = [.5, 1, 1.5, 2, 2.5, 3]
speed = [100, 150, 120, 240, 260, 189]

plt.figure(figsize=[6, 6])
plt.plot(time, speed, color='darkcyan', marker='o', linestyle='-', linewidth=2, label='Bike Speed/s')
plt.title('Time-Wise Moto Bike Speed Fluctuation', fontsize=15)
plt.legend(loc='upper left')
plt.grid(color='g', linewidth=.5, linestyle='-.')
# plt.xlim(.5, 6)
# plt.ylim(100, 500)
plt.xlabel("Time", color='b')
plt.ylabel("Speed", color='violet')
plt.show()
```

# **Bar Graph / Bar Chart**

```python
months = ['Jan', 'Feb', 'Mar', 'April', 'May', 'June']
sales = [200, 1500, 700, 350, 1700, 2500]

# plt.bar(months, sales, width=.5, color='yellow', edgecolor='black', linewidth=2)
plt.bar(months, sales, width=.5, color=['yellow', 'green', 'blue', 'red', 'orange', 'violet'], edgecolor='black', linewidth=2)
plt.xlabel("Months")
plt.ylabel("Sales/Mon", rotation=0)
plt.show()
```
