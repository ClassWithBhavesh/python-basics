```python
import pandas as pd
import numpy as np


prodDf = pd.DataFrame({
    "product_name": ['ethnic kurti', 'mens joggers', 'Womens necklace', 'Womens perfume', 'Mens perfume', 'womens watches', 'pen tab'],
    "product_company_name": ['SlayVana', 'Souled Store', 'Shiver Jewellers', 'Bella Vita', 'Park Avenue', 'Titans', 'wacom'],
    "product_price": [799, None, 499, np.nan, 450, None, 2499]
})

prodDf
```

### to find whether our dataframe is having null, nan or none value we'll use .**isnull()**

```python
prodDf.isnull().sum()

# print(int(True))
```

### To fill the null, nan or none values with some default values we'll use .***fillna()***

```python
# prodDf.isnull().sum()

# filling null, nan or none value with a default value
newDf = prodDf.fillna(349, inplace=False) # by default inplace is False everywhere, if we want to change the existing dataframe then we've to provide inplace=True

prodDf.isnull().sum()
prodDf
newDf
```
