# ***Filtering values in Pandas***


```python
import pandas as pd
import numpy as np


prodDf = pd.DataFrame({
    "product_name": ['ethnic kurti', 'mens joggers', 'Womens necklace', 'Womens perfume', 'Mens perfume', 'womens watches', 'pen tab'],
    "product_company_name": ['SlayVana', 'Souled Store', 'Shiver Jewellers', 'Bella Vita', 'Park Avenue', 'Titans', 'wacom'],
    "product_price": [799, None, 499, np.nan, 450, None, 2499]
})

# prodDf.fillna(1200, inplace=True)


prodDf

# filtering dataframe entried whose product price is greater than 1200
# prodDf[prodDf['product_price'] > 1200]

prodDf['final_price_after_discount'] = prodDf['product_price'] - (prodDf['product_price'] * 15/100)
prodDf

# sizeCol = pd.Series(['XL', "XXL", "M", '120ml', '150ml', 'L', '7x5 inches'])

# sizeCol

# prodDf.insert(2, 'product_size', sizeCol)
# prodDf

# removing a single column
prodDf.drop(index=2)
prodDf


# removing multiple records (rows)
prodDf.drop([0,2])


# removing a single column by it's name
prodDf.drop('final_price_after_discount', axis=1)



# removing or droping null, nan or none values from our DataFrame
# isnull()
# fillna()
# dropna()

# prodDf.isnull().sum()
prodDf.dropna()
```
